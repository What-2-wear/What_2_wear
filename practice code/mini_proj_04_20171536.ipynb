```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import cross_val_score
from  sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import classification_report
from sklearn.ensemble import VotingClassifier, BaggingClassifier, RandomForestClassifier
```

```python
df = pd.read_csv("../datasets/winequality-red.csv")
```

```python
df.head(10)
```

```python
df.describe()
```

```python
df = df.sample(frac=1)
```

```python
X, y = df.drop("quality", axis=1),  df["quality"]
```

```python
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.metrics import mean_squared_error
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.8)
```

```python
model = ExtraTreesRegressor(max_depth=20, max_features=9)
```

```python
model.fit(X_train, y_train)
```

```python
predicted = model.predict(X_test)
train_predicted = model.predict(X_train)
train_mse = mean_squared_error(y_train, train_predicted)
mse = mean_squared_error(y_test, predicted)
rmse = np.sqrt(mse)
train_rmse = np.sqrt(train_mse)
print(train_rmse, rmse)
```

```python
def displayscores(scores):
    print("Scores: ", scores)
    print("Mean: ", scores.mean())
    print("Standard deviation: ", scores.std())
```

```python
scores = cross_val_score(model, X, y, scoring="neg_mean_squared_error", n_jobs = -1, cv=10)
displayscores(np.sqrt(-scores))
```
